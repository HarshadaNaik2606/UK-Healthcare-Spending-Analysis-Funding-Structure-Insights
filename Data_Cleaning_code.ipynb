# UK Healthcare Spending Trends & Funding Structure Analysis – Data Cleaning

## Project Overview
This notebook documents the data cleaning and preparation process for the **UK Healthcare Spending Trends & Funding Structure Analysis** project.

The dataset was sourced from the **Office for National Statistics (ONS)** and contains historical UK healthcare expenditure data from **1997 to 2012**, including:

- Total healthcare expenditure
- Public sector healthcare expenditure
- Private sector healthcare expenditure
- Revision values

## Objective
The purpose of this notebook is to clean and structure the raw dataset so it can be used for:

- exploratory data analysis
- SQL-based analysis
- Power BI dashboard development

## Data Cleaning Tasks
The notebook performs the following steps:

1. Load the raw Excel dataset
2. Identify and extract the relevant table
3. Rename and standardise columns
4. Convert values to numeric format
5. Create derived analytical fields such as:
   - Public funding share (%)
   - Private funding share (%)
   - Year-on-year growth (%)
6. Export the cleaned dataset for downstream analysis

## Tools Used
- Python
- Pandas
- NumPy

```python
import pandas as pd
import numpy as np
```

```python
file_path = "referencetable10revisionstototalpublicsectorandprivatesectorhealthcareexpenditureintheuk19972012_tcm77-399797.xls"

df_raw = pd.read_excel(file_path, sheet_name=0, header=None, engine="xlrd")
df_raw.head(15)
```

```python
print("Raw shape:", df_raw.shape)
```

```python
df = df_raw.copy()

# keep rows where first column looks like a year
df = df[df[0].apply(lambda x: isinstance(x, (int, float)) and 1900 <= x <= 2100)]

df.head()
```

```python
df = df[[0, 1, 2, 3, 5, 6, 7, 9,10,11]].copy()
df.head()
```

```python
df.columns = [
    "year",
    "total_2015",
    "total_2014",
    "total_revision",
    "public_2015",
    "public_2014",
    "public_revision",
    "private_2015",
    "private_2014",
    "private_revision"
]

df.head()
```

```python
df
```

```python
for col in df.columns:
    df[col] = pd.to_numeric(df[col], errors="coerce")

df.info()
```

```python
df = df.sort_values("year").reset_index(drop=True)
df.head()
```

```python
df["public_share_pct_2015"] = (df["public_2015"] / df["total_2015"]) * 100
df["private_share_pct_2015"] = (df["private_2015"] / df["total_2015"]) * 100

df["total_yoy_growth_pct_2015"] = df["total_2015"].pct_change() * 100
df["public_yoy_growth_pct_2015"] = df["public_2015"].pct_change() * 100
df["private_yoy_growth_pct_2015"] = df["private_2015"].pct_change() * 100
```

```python
df["total_yoy_growth_pct_2015"] = df["total_yoy_growth_pct_2015"].fillna(0)
df["public_yoy_growth_pct_2015"] = df["public_yoy_growth_pct_2015"].fillna(0)
df["private_yoy_growth_pct_2015"] = df["private_yoy_growth_pct_2015"].fillna(0)
```

The first year (1997) has no previous year available for comparison, so the year-on-year growth calculation returns NaN. This missing value was replaced with 0 to maintain a complete dataset and ensure consistency for further analysis in SQL and Power BI.

```python
df = df.round(2)
df.head()
```

```python
latest_year = df["year"].max()
latest_total = df.loc[df["year"] == latest_year, "total_2015"].values[0]
latest_public_share = df.loc[df["year"] == latest_year, "public_share_pct_2015"].values[0]
highest_growth = df["total_yoy_growth_pct_2015"].max()
highest_growth_year = df.loc[df["total_yoy_growth_pct_2015"] == highest_growth, "year"].values[0]

print("Latest year:", latest_year)
print("Latest healthcare spend (£bn):", latest_total)
print("Public funding share in latest year (%):", latest_public_share)
print("Highest YoY growth (%):", highest_growth)
print("Highest growth year:", highest_growth_year)
```

```python
df.to_csv("cleaned_healthcare_expenditure.csv", index=False)
df.to_excel("cleaned_healthcare_expenditure.xlsx", index=False)

print("Cleaned dataset exported successfully.")
```
